```python
neworder = ['Rain_OR_SNOW', 'Temperature (C)',
       'Apparent Temperature (C)', 'Humidity', 'Wind Speed (km/h)',
       'Wind Bearing (degrees)', 'Visibility (km)', 'Pressure (millibars)',
       'Condensation', 'Solar irradiance intensity','Cloud_Condition']
train=train.reindex(columns=neworder)
```

```python
train.head()
```

```python
train['Rain_OR_SNOW'].fillna(train['Rain_OR_SNOW'].mode()[0],inplace=True)
```

```python
train['Rain_OR_SNOW'].mode()[0]
```

```python
train["Temperature (C)"] = pd.to_numeric(train["Temperature (C)"], errors='coerce')
```

```python
#replace the missing values with mode values
for value in ['Temperature (C)','Apparent Temperature (C)','Humidity','Wind Speed (km/h)','Wind Bearing (degrees)','Visibility (km)','Pressure (millibars)']:
    train[value].fillna(train[value].mean(),inplace=True)
```

```python
train.isnull().sum()
```

```python
colname=[]
for x in train.columns:
    if train[x].dtype=='object':
        colname.append(x)
colname
```

```python
# For preprocessing the data
from sklearn import preprocessing

le=preprocessing.LabelEncoder()

#for x in colname:
#    train[x]=le.fit_transform(train[x])
```

```python
train.head()
```

```python
for x in colname:
    train[x]=le.fit_transform(train[x])
    le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
    print('Feature', x)
    print('mapping', le_name_mapping)
```

```python
X = train.values[:,:-1]
Y = train.values[:,-1]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X)

X = scaler.transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split

#Split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3,
random_state=10)
```

```python
#predicting using the Decision_Tree_Classifier
from sklearn.tree import DecisionTreeClassifier

model_DecisionTree=DecisionTreeClassifier(random_state=10)

#fit the model on the data and predict the values
model_DecisionTree.fit(X_train,Y_train)
Y_pred=model_DecisionTree.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,\
classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```
